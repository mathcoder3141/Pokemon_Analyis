```python
import pandas as pd
pd.set_option("display.max_columns", None) 
# pd.set_option("display.max_rows", None)
# Uncomment the above if you want to view the full data
pokemon = pd.read_csv('pokemon.csv')
```

```python
pokemon.columns
```

```python
column_order = ['name', 'japanese_name', 'pokedex_number', 'percentage_male', 'type1', 'type2', 'classfication', 'height_m',
               'weight_kg', 'capture_rate', 'base_egg_steps', 'abilities', 'experience_growth', 'base_happiness', 'against_bug', 
               'against_dark', 'against_dragon', 'against_electric', 'against_fairy', 'against_fight', 'against_fire',
               'against_flying', 'against_ghost', 'against_grass', 'against_ground', 'against_ice', 'against_normal', 
               'against_poison', 'against_psychic', 'against_rock', 'against_steel', 'against_water', 'base_total', 'hp', 'attack',
               'defense', 'sp_attack', 'sp_defense', 'speed', 'generation', 'is_legendary']
pokemon = pokemon[column_order]
pokemon = pokemon.rename({"classfication":"classification"})
```

To make it easier to explore data in a specific generation, let's subset our data such that each row is from a single generation.

```python
gen_1 = pokemon[pokemon["generation"] == 1]
gen_2 = pokemon[pokemon["generation"] == 2]
gen_3 = pokemon[pokemon["generation"] == 3]
gen_4 = pokemon[pokemon["generation"] == 4]
gen_5 = pokemon[pokemon["generation"] == 5]
gen_6 = pokemon[pokemon["generation"] == 6]
gen_7 = pokemon[pokemon["generation"] == 7]
```

```python
for frame in (gen_1, gen_2, gen_3, gen_4, gen_5, gen_6, gen_7):
    print(frame.shape[0])
```

It looks like the subsetting of generation worked correctly!

Since we're looking at legendary Pokémon, let's take a look at how the number of new legendary Pokémon changes over time. To do this, let's modify the above statement to reflect that we want each legendary Pokémon from it's respective generation in it's own dataframe.

```python
gen_1_legendary = pokemon[(pokemon["generation"] == 1) & (pokemon["is_legendary"] == 1)]
gen_2_legendary = pokemon[(pokemon["generation"] == 2) & (pokemon["is_legendary"] == 1)]
gen_3_legendary = pokemon[(pokemon["generation"] == 3) & (pokemon["is_legendary"] == 1)]
gen_4_legendary = pokemon[(pokemon["generation"] == 4) & (pokemon["is_legendary"] == 1)]
gen_5_legendary = pokemon[(pokemon["generation"] == 5) & (pokemon["is_legendary"] == 1)]
gen_6_legendary = pokemon[(pokemon["generation"] == 6) & (pokemon["is_legendary"] == 1)]
gen_7_legendary = pokemon[(pokemon["generation"] == 7) & (pokemon["is_legendary"] == 1)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

ax = sns.pointplot(x=[x for x in range(1, 8)], y=[gen_1_legendary.shape[0], gen_2_legendary.shape[0], gen_3_legendary.shape[0], gen_4_legendary.shape[0], 
                                                 gen_5_legendary.shape[0], gen_6_legendary.shape[0], gen_7_legendary.shape[0]])
ax.set(xlabel="Generation", ylabel="Number of New Legendaries")
plt.show()
```

From this graph, we can see that the number of legendary Pokémon introduced is constant between generation two and generation six as well as constant between generation four and generation five. From a player's standpoint, I'm glad they kept the number of new legendary Pokémon under 10 for four out of the seven generations. I wonder; however, if the trend for stats of legendary Pokémon are positive or negative i.e., if they got weaker or stronger. Let's explore this!

```python
stats_cols = ["name", "hp", "attack", "defense", "sp_attack", "sp_defense", "speed", "base_total"]
gen_1_legendary_stats = gen_1_legendary[stats_cols]
gen_2_legendary_stats = gen_2_legendary[stats_cols]
gen_3_legendary_stats = gen_3_legendary[stats_cols]
gen_4_legendary_stats = gen_4_legendary[stats_cols]
gen_5_legendary_stats = gen_5_legendary[stats_cols]
gen_6_legendary_stats = gen_6_legendary[stats_cols]
gen_7_legendary_stats = gen_7_legendary[stats_cols]
```

Eyeballing stats isn't a good way of spotting an immediate trend. Let's look at some more pretty graphs! It's important to note that with each evolution, a Pokémon will naturally get stronger. With that being said, Cosmog and Cosmoem won't be included in this since their final evolutions are Solgaleo or Lunala. We'll drop these rows and identify a trend using the mean of the base totals of all the legendary Pokémon throughout the generations.

```python
gen_1_legendary_mean_stats = gen_1_legendary_stats.mean()
gen_2_legendary_mean_stats = gen_2_legendary_stats.mean()
gen_3_legendary_mean_stats = gen_3_legendary_stats.mean()
gen_4_legendary_mean_stats = gen_4_legendary_stats.mean()
gen_5_legendary_mean_stats = gen_5_legendary_stats.mean()
gen_6_legendary_mean_stats = gen_6_legendary_stats.mean()
gen_7_legendary_mean_stats = gen_7_legendary_stats.drop([788, 789], axis=0).mean()

ax = sns.pointplot(x=[x for x in range(1, 8)], y=[gen_1_legendary_mean_stats.mean(), gen_2_legendary_mean_stats.mean(), gen_3_legendary_mean_stats.mean(), 
                                             gen_4_legendary_mean_stats.mean(), gen_5_legendary_mean_stats.mean(), gen_6_legendary_mean_stats.mean(), 
                                             gen_7_legendary_mean_stats.mean()])
ax.set(xlabel="Generation", ylabel="Avg Base Stats")
plt.show()
```

Eh ... the trend doesn't seem too bad. The average stats total tends the generally hover between 160 and about 200. Now let's see which legendary Pokémon are the strongest and weakest. Again, we'll drop Cosmog and Cosmoem.

To get towards the end of this analysis, I would like to examine the top five Pokémon of each unique type in the `type1` field. I would like to do this to try to get a good Pokémon team would be if I were to compete across WFC or start getting into Pokemeon VGC. For that reason, I will be filtering down the `pokemon` dataframe down to a dataframe that contains each unique type along with those that aren't legendary since most of them are in the Über tier or outright banned from WFC or Pokemon VGC.

```python
grass_pokemon_not_legendary = pokemon[(pokemon["type1"] == "grass") & (pokemon["is_legendary"] == 0)]
fire_pokemon_not_lengendary = pokemon[(pokemon["type1"] == "fire") & (pokemon["is_legendary"] == 0)]
water_pokemon_not_legendary = pokemon[(pokemon["type1"] == "water") & (pokemon["is_legendary"] == 0)]
bug_pokemon_not_legendary = pokemon[(pokemon["type1"] == "bug") & (pokemon["is_legendary"] == 0)]
normal_pokemon_not_legendary = pokemon[(pokemon["type1"] == "normal") & (pokemon["is_legendary"] == 0)]
poison_pokemon_not_legendary = pokemon[(pokemon["type1"] == "poison") & (pokemon["is_legendary"] == 0)]
electric_pokemon_not_legendary = pokemon[(pokemon["type1"] == "electric") & (pokemon["is_legendary"] == 0)]
ground_pokemon_not_legendary = pokemon[(pokemon["type1"] == "ground") & (pokemon["is_legendary"] == 0)]
fairy_pokemon_not_legendary = pokemon[(pokemon["type1"] == "fairy") & (pokemon["is_legendary"] == 0)]
fighting_pokemon_not_legendary = pokemon[(pokemon["type1"] == "fighting") & (pokemon["is_legendary"] == 0)]
psychic_pokemon_not_legendary = pokemon[(pokemon["type1"] == "psychic") & (pokemon["is_legendary"] == 0)]
rock_pokemon_not_legendary = pokemon[(pokemon["type1"] == "rock") & (pokemon["is_legendary"] == 0)]
ghost_pokemon_not_legendary = pokemon[(pokemon["type1"] == "ghost") & (pokemon["is_legendary"] == 0)]
ice_pokemon_not_legendary = pokemon[(pokemon["type1"] == "ice") & (pokemon["is_legendary"] == 0)]
dragon_pokemon_not_legendary = pokemon[(pokemon["type1"] == "dragon") & (pokemon["is_legendary"] == 0)]
dark_pokemon_not_legendary = pokemon[(pokemon["type1"] == "dark") & (pokemon["is_legendary"] == 0)]
steel_pokemon_not_legendary = pokemon[(pokemon["type1"] == "steel") & (pokemon["is_legendary"] == 0)]
flying_pokemon_not_legendary = pokemon[(pokemon["type1"] == "flying") & (pokemon["is_legendary"] == 0)]
```

Now that we created multiple sub-dataframes that meet our requirements, let's display them to try to form a super strong Pokémon team that contain non-legendaries to meet our original goal. To do this, we'll sort the values by `base_total`, and since we love to be on the offense, we'll also sort by `attack` so that if two Pokémon have equal values for `base_total`, the Pokémon with the greater `attack` stat will get listed first. For sanity sake and to check if the dataframe logic is correct, we'll only include the following columns: `name`, `type1`, `base_total`, and `attack`.

```python
top_five_grass = grass_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_fire = fire_pokemon_not_lengendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_water = water_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_bug = bug_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_normal = normal_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_poison = poison_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_electric = electric_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_ground = ground_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_fairy = fairy_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_fighting = fighting_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_psychic = psychic_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_rock = rock_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_ghost = ghost_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_ice = ice_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_dragon = dragon_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_dark = dark_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_steel = steel_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
top_five_flying = flying_pokemon_not_legendary.sort_values(by=['base_total', 'attack'], ascending=False)[['name', 'type1', 'base_total', 'attack']].head()
```

Now that we created multiple sub-dataframes that meet our requirements, let's display them to try to form a offensive Pokémon team that contain non-legendaries to meet our original goal.

```python
for frame in [top_five_grass, top_five_fire, top_five_water, top_five_bug, top_five_normal, top_five_poison, top_five_electric, top_five_ground, 
              top_five_fairy, top_five_fighting, top_five_psychic, top_five_rock, top_five_ghost, top_five_ice, top_five_dragon, top_five_dark, 
              top_five_steel, top_five_flying]:
    print(frame)
```

```python
pokemon.drop(770, inplace=True)
pokemon.drop(771, inplace=True)
pokemon.drop(145, inplace=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(pokemon[['base_egg_steps', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']], pokemon['is_legendary'])
```

```python
from IPython.display import Image  
from sklearn import tree
import pydotplus

dot_data = tree.export_graphviz(clf, out_file=None, 
                                feature_names=['base_egg_steps', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed'],  
                                class_names=['0', '1'])

# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)  
```

```python
# Write graph to file
graph.write_png('pokemon.png');
```

```python
if 1 in clf.predict([[1250000, 95, 95, 95, 95, 95, 59]]):
    print("Type: Null is a legendary")
else:
    print("Type: Null is not legendary")
    
if 1 in clf.predict([[1250000, 95, 95, 95, 95, 95, 95]]):
    print("Silvally is a legendary")
else:
    print("Silvally is not legendary")
```
