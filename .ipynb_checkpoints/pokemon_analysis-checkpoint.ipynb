```python
import pandas as pd
# pd.set_option("display.max_columns", None) 
# pd.set_option("display.max_rows", None)
# Uncomment the above if you want to view the full data
pokemon = pd.read_csv('pokemon.csv')
```

```python
pokemon.columns
```

```python
pokemon = pokemon.rename({"classfication":"classification"})
```

* name: The English name of the Pokemon
* japanese_name: The Original Japanese name of the Pokemon
* pokedex_number: The entry number of the Pokemon in the National Pokedex
* percentage_male: The percentage of the species that are male. Blank if the Pokemon is genderless.
* type1: The Primary Type of the Pokemon
* type2: The Secondary Type of the Pokemon
* classification: The Classification of the Pokemon as described by the Sun and Moon Pokedex
* height_m: Height of the Pokemon in metres
* weight_kg: The Weight of the Pokemon in kilograms
* capture_rate: Capture Rate of the Pokemon
* base_egg_steps: The number of steps required to hatch an egg of the Pokemon
* abilities: A stringified list of abilities that the Pokemon is capable of having
* experience_growth: The Experience Growth of the Pokemon
* base_happiness: Base Happiness of the Pokemon
* against_?: Eighteen features that denote the amount of damage taken against an attack of a particular type
* hp: The Base HP of the Pokemon
* attack: The Base Attack of the Pokemon
* defense: The Base Defense of the Pokemon
* sp_attack: The Base Special Attack of the Pokemon
* sp_defense: The Base Special Defense of the Pokemon
* speed: The Base Speed of the Pokemon
* generation: The numbered generation which the Pokemon was first introduced
* is_legendary: Denotes if the Pokemon is legendary.

Since we're looking at legendary pokemon, let's take a look at how the number of new legendary pokemon changes over time. To do this, let's modify the above statement to reflect that we want each legendary pokemon from it's respective generation in it's own dataframe.

```python
gen_1_legendary = pokemon[(pokemon["generation"] == 1) & (pokemon["is_legendary"] == 1)]
gen_2_legendary = pokemon[(pokemon["generation"] == 2) & (pokemon["is_legendary"] == 1)]
gen_3_legendary = pokemon[(pokemon["generation"] == 3) & (pokemon["is_legendary"] == 1)]
gen_4_legendary = pokemon[(pokemon["generation"] == 4) & (pokemon["is_legendary"] == 1)]
gen_5_legendary = pokemon[(pokemon["generation"] == 5) & (pokemon["is_legendary"] == 1)]
gen_6_legendary = pokemon[(pokemon["generation"] == 6) & (pokemon["is_legendary"] == 1)]
gen_7_legendary = pokemon[(pokemon["generation"] == 7) & (pokemon["is_legendary"] == 1)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

ax = sns.pointplot(x=[x for x in range(1, 8)], y=[gen_1_legendary.shape[0], gen_2_legendary.shape[0], gen_3_legendary.shape[0], gen_4_legendary.shape[0], 
                                                 gen_5_legendary.shape[0], gen_6_legendary.shape[0], gen_7_legendary.shape[0]])
ax.set(xlabel="Generation", ylabel="Number of New Legendaries")
plt.show()
```

From this graph, we can see that the number of legendary pokemon introduced is constant between generation two and generation six as well as constant between generation four and generation five. From a player's standpoint, I'm glad they kept the number of new legendary pokemon under 10 for four out of the seven generations. I wonder; however, if the trend for stats of legendary pokemon are positive or negative i.e., if they got weaker or stronger. Let's explore this!

```python
stats_cols = ["name", "hp", "attack", "defense", "sp_attack", "sp_defense", "speed", "base_total"]
gen_1_legendary_stats = gen_1_legendary[stats_cols]
gen_2_legendary_stats = gen_2_legendary[stats_cols]
gen_3_legendary_stats = gen_3_legendary[stats_cols]
gen_4_legendary_stats = gen_4_legendary[stats_cols]
gen_5_legendary_stats = gen_5_legendary[stats_cols]
gen_6_legendary_stats = gen_6_legendary[stats_cols]
gen_7_legendary_stats = gen_7_legendary[stats_cols]
```

Eyeballing stats isn't a good way of spotting an immediate trend. Let's look at some more pretty graphs! It's important to note that with each evolution, a pokemon will naturally get stronger. With that being said, Cosmog and Cosmoem won't be included in this since their final evolutions are Solgaleo or Lunala. We'll drop these rows and identify a trend using the mean of the base totals of all the legendary pokemon throughout the generations.

```python
gen_1_legendary_mean_stats = gen_1_legendary_stats.mean()
gen_2_legendary_mean_stats = gen_2_legendary_stats.mean()
gen_3_legendary_mean_stats = gen_3_legendary_stats.mean()
gen_4_legendary_mean_stats = gen_4_legendary_stats.mean()
gen_5_legendary_mean_stats = gen_5_legendary_stats.mean()
gen_6_legendary_mean_stats = gen_6_legendary_stats.mean()
gen_7_legendary_mean_stats = gen_7_legendary_stats.drop([788, 789], axis=0).mean()

ax = sns.pointplot(x=[x for x in range(1, 8)], y=[gen_1_legendary_mean_stats.mean(), gen_2_legendary_mean_stats.mean(), gen_3_legendary_mean_stats.mean(), 
                                             gen_4_legendary_mean_stats.mean(), gen_5_legendary_mean_stats.mean(), gen_6_legendary_mean_stats.mean(), 
                                             gen_7_legendary_mean_stats.mean()])
ax.set(xlabel="Generation", ylabel="Avg Base Stats")
plt.show()
```

Eh ... the trend doesn't seem too bad as the all tend the generally hover between 160 and about 200. Now let's see which legendary pokemon are the strongest and weakest. Again, we'll drop Cosmog and Cosmoem.
